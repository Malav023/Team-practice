```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
outcome_values = np.array([0, 1, 1, 2, 3, 4, 4, 4, 5, 6, 6, 6, 6])

# 1. Count occurrences of each value
# We can use np.bincount to efficiently count occurrences for non-negative integer arrays
# The minlength ensures the array covers all values from 0 to 6, even if some are missing in the data
counts = np.bincount(outcome_values, minlength=7) # Range is 0 to 6 (7 possible outcomes)

# 2. Normalize counts to get probabilities
total_outcomes = len(outcome_values)
probability_array = counts / total_outcomes

# results
print(f"Outcome Counts: {counts}")
print(f"Probability Array (PMF): {probability_array}")
```

```python
outcomes = np.arange(7) # [0, 1, 2, 3, 4, 5, 6]

# Create the custom discrete distribution
custom_dist = stats.rv_discrete(name='custom_empirical_dist', values=(outcomes, probability_array))

# You can now use SciPy functions with your custom distribution:
print(f"Probability of outcome 4: {custom_dist.pmf(4)}")
print(f"Mean of the distribution: {custom_dist.mean()}")
print(f"Variance of the distribution: {custom_dist.var()}")
```

```python
plt.plot(outcomes,probability_array,'o')
plt.grid(True)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 25
p = 0.85

# Possible outcomes
x = np.arange(0, n + 1)

# Binomial PMF
pmf = binom.pmf(x, n, p)

# Plot
plt.figure(figsize=(8, 5))

# Vertical lines
plt.vlines(x, 0, pmf)

# Circles at the top of each line
plt.plot(x, pmf, 'o')

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution n=25, p=0.85")
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Fixed number of trials
n = 100

# x-values
x = np.arange(0, n + 1)

# Probability values
p_values = np.arange(0.1, 1.0, 0.1)

# Create 3x3 subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 10))

# Flatten axes array for easy looping
axes = axes.flatten()

# Plot each PMF
for i, p in enumerate(p_values):
    pmf = binom.pmf(x, n, p)
    
    # Vertical lines
    axes[i].vlines(x, 0, pmf)
    
    # Dots at top
    axes[i].plot(x, pmf)
```

## Question 4

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Fixed probability
p = 0.5

# Different values of n
n_values = np.arange(10, 100, 10)

# Create 3x3 subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 10))

# Flatten axes for easy indexing
axes = axes.flatten()

# Plot PMFs
for i, n in enumerate(n_values):
    x = np.arange(0, n + 1)
    pmf = binom.pmf(x, n, p)
    
    # Vertical lines
    axes[i].vlines(x, 0, pmf)
    
    # Dots at top
    axes[i].plot(x, pmf, 'o', markersize=3)
    
    axes[i].set_title(f"binomial(n={n}, p=0.5)")
    axes[i].set_xlabel("Successes")
    axes[i].set_ylabel("Probability")

# Adjust layout
plt.tight_layout()

# Display grid
plt.show()
```
