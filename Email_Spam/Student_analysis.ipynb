```python
hours_per_day=[1.3,2.5,1.0,3.0]
marks_obtained=[20,25,17,28]
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
hours_per_day_df=pd.DataFrame(hours_per_day)
marks_obtained_df=pd.DataFrame(marks_obtained)
```

```python
plt.scatter(hours_per_day_df,marks_obtained_df)
plt.show()
```

```python
linr=LinearRegression()

linr.fit(hours_per_day_df,marks_obtained_df)
```

```python
y_pred=linr.predict(hours_per_day_df)
```

```python
plt.scatter(hours_per_day_df,marks_obtained_df)
plt.plot(hours_per_day_df,y_pred)
plt.show()
```
