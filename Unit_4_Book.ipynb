```python
#generating datapoints for testing 

import numpy as np 

rng=np.random.default_rng(seed=42)
m=200
x=2*rng.random((m,1))
y=4+3*x+rng.standard_normal((m,1))
```

```python
#plotting generated data-points
import matplotlib.pyplot as plt 

plt.scatter(x,y)
```

```python
#calculating theta using equation 

from sklearn.preprocessing import add_dummy_feature

x_b=add_dummy_feature(x)
theta=np.linalg.inv(x_b.T @ x_b) @ x_b.T @ y
theta 
```

```python
x_new=np.array([[0],[2]])
# x_new.shape
x_new_b=add_dummy_feature(x_new)
x_new_b

y_pred=x_new_b@theta
y_pred
```

```python
plt.plot(x_new,y_pred,'r-')
plt.plot(x,y,"b.")
```

## Batch Gradient Descent

```python
eta=0.1 #learning rate
n_epochs=1000 #no. of times to iterate over training set
m=len(x_b) #no. of instances

rng=np.random.default_rng(seed=42)
theta=rng.standard_normal((2,1))

#Gradient vector of the cost function (gradient variable)
for epoch in range(n_epochs):
    gradient=2/m*x_b.T@(x_b@theta-y)
    theta=theta-eta*gradient
```
